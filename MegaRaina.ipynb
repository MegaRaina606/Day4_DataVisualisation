```python
import numpy as np
import pandas as pd
import seaborn as sns

import matplotlib as mpl
import matplotlib.pyplot as plt
%matplotlib inline
```

```python
from numpy.random import randn, randint, uniform, sample
```

```python
#4.Bar Plot
x=['Jan','Feb','March','Apr','May','June','July','Aug','Sep','Oct','Nov','Dec']
y=['15k','25k','55k','70k','82k','1l','1.8l','3.2l','5.5l','12l','15l','19l']


plt.bar(x,y,label="Covid active cases ")
plt.title('Bar graph')
plt.ylabel('no. of active cases')
plt.xlabel('Month')
plt.legend()
plt.show()
```

```python
#5.Line Chart
x=['Jan','Feb','March','Apr','May','June','July','Aug','Sep','Oct','Nov','Dec']
y=['1k','5k','7k','8k','12k','16k','21k','50k','75k','2l','7l','10l']

plt.title("Line Plot")
plt.xlabel("Month")
plt.ylabel("Active Covid Recovery Rates")
plt.plot(x,y)
```

```python
#3.Pie Chart
labels = ['Hyunda','Ford','Maruti Suzuki','MG Motor','Kia','Toyota']
sizes = [29,22,112,22,60,30]
colors = ['gold','yellowgreen','lightcoral','lightskyblue','Red','Blue']

#Plot
plt.pie(sizes,labels=labels,colors=colors,
autopct='%1.2f%%',shadow=True)

plt.axis('equal')
plt.show()
```
